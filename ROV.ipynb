```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Step 1: Create a graph with position and distances
position = ['H', 'AD', 'TOP', 'MID', 'JU', 'SUB']
distances = {
    ('H', 'AD'): 10,
    ('H', 'TOP'): 15,
    ('H', 'MID'): 20,
    ('H', 'JU'): 25,
    ('H', 'SUB'): 10
}

# Create a directed graph
G = nx.DiGraph()

# Add nodes (position)
G.add_nodes_from(position)

# Add weighted edges (distances)
for (point1, point2), distance in distances.items():
    G.add_edge(point1, point2, weight=distance)
    # Note: Only add edge in one direction to keep it directed

# Load the background image
img = mpimg.imread('ROV_map.jpg')  # Replace 'ROV_map.jpg' with your image path

# Define positions of the nodes manually to fit the image
pos = {
    'H': (-0.67, -0.67),  #(v,h)
    'AD': (0.57, -0.5),
    'TOP': (-0.4, 0.55),
    'MID': (0.02, 0.1),
    'JU': (-0.35, 0.1),
    'SUB': (0.1, 0)
}

# Create the plot
plt.figure(figsize=(10, 8))

# Display the image
plt.imshow(img, extent=[-1, 1, -1, 1], aspect='auto')

# Draw the graph
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, edge_color='black', arrowsize=20, connectionstyle='arc3, rad=0')
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f'{d["weight"]}' for u, v, d in G.edges(data=True)}, font_color='red')
plt.title("Map Rov")
plt.show()
```
